# Embeddings reducidos

```python
from openai import OpenAI
import numpy as np


def normalize_l2(x):
    x = np.array(x)
    if x.ndim == 1:
        norm = np.linalg.norm(x)
        if norm == 0:
            return x
        return x / norm
    else:
        norm = np.linalg.norm(x, 2, axis=1, keepdims=True)
        return np.where(norm == 0, x, x / norm)


response = client.embeddings.create(
    model="text-embedding-3-large", input="Sigueme en youtube y apoya el canal", encoding_format="float"
)

origina_dim = len(response.data[0].embedding)
cut_dim = response.data[0].embedding[:256]
norm_dim = normalize_l2(cut_dim)

print('La dimensionalidad original, es:',origina_dim,' y la nueva dimensionalidad es ahora de:',  len(norm_dim))
```

# Creemos un buscador semantico

```python
def get_embedding(text, model="text-embedding-3-small", azure_endpoint=False):
   text = text.replace("\n", " ")
   if azure_endpoint:
    model = 'embeddings_data'
    return client_azure.embeddings.create(input=[text],model= model).data[0].embedding
   else:
    return client.embeddings.create(input = [text], model=model).data[0].embedding
```

### Limpiando datos

```python
import pandas as pd

df = pd.read_csv('/content/Netflix TV Shows and Movies.csv')
```

```python
df.head(5)
```

```python
df.drop(['index','id'],axis=1,inplace=True)
```

```python
df['total_text'] = df.apply(lambda x: str(x['title']) + ' ' + str(x['description']), axis=1)
```

```python
df['text_embedding'] = df['total_text'].apply(lambda x: get_embedding(x, model='text-embedding-3-small'))
```

```python
df.head(5)
```

# Search

```python
from sklearn.metrics.pairwise import cosine_similarity

def top_n_similar_movies(query, n, model='text-embedding-3-small', azure_endpoint=False):
    # Genera el embedding para la consulta
    query_embedding = get_embedding(query,model=model,azure_endpoint=azure_endpoint)
    query_embedding = np.array(query_embedding).reshape(1, -1)
    # Calcula la similitud del coseno entre la consulta y todas las películas
    similarities = cosine_similarity(query_embedding, np.array(df['text_embedding'].tolist()))

    # Obtiene los índices de los n resultados más similares
    top_n_indices = similarities[0].argsort()[-n:][::-1]
    # Devuelve las n resultados más similares
    return df.iloc[top_n_indices][['title', 'description','type','imdb_id', 'imdb_score', 'imdb_votes']]
```

```python
top_n_similar_movies('una historia de viajes en el tiempo',3)
```
